```python
import numpy as np
from numpy import pi, cos, sin
from numpy.linalg import lstsq
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns

sns.set_context("poster")
sns.set_style("whitegrid") 
# NOTE: the following lines might need to be disabled if you get errors related to LaTeX!
#rc('font',**{'family':'serif','serif':['Computer Modern Roman']})
rc("text",usetex=True)
rc('lines', linewidth=2.5)
```

## Loading the data

```python
cardiac = np.loadtxt("cardiac_true.txt")
fs = 1e3 # sampling rate [Hz]
t = np.arange(len(cardiac))/fs # time vector for the cardiac signal
t_end = 5.0 # in seconds
ff = np.arange(0,1.5,1e-2) # frequencies to be considered
Y = cardiac[t<t_end] # sampled time range
Y = Y - Y.mean() # centering data
T = t[t<t_end] # time vector for cut sample
```

## Fitting a Fourier series with 3 harmonics

```python
min_ind = 0
min_err = 0
T = T.reshape(-1,1) # make T to a column vector 
# for each frequency in the range
for i,f in enumerate(ff):
    # for only three harmonics (np.arange(1,4) = [1,2,3])
    #X = np.concatenate([cos(2*pi*T[:,np.newaxis]*np.arange(1,4)*f), sin(2*pi*T[:,np.newaxis]*np.arange(1,4)*f)],axis=1)
    X = np.concatenate([cos(2*pi*T*np.arange(1,4)*f), sin(2*pi*T*np.arange(1,4)*f)],axis=1)
    c = lstsq(X,Y,rcond=None)[0]
    err = np.mean((Y-X@c)**2)
    # if the index is zero or the error is smaller than the smallest measured error
    if min_ind==0 or err < min_err:
        # update parameters
        min_ind = i
        min_err = err
        f_best = f
        c_best = c
        X_best = X
```

## Plotting the cardiac signal and the Fourier fit

```python
plt.figure(figsize=(10,6))
plt.plot(T, X_best@c_best)
plt.plot(T, Y, "--")
plt.legend(["fourier fit", "signal"])
plt.yticks(np.arange(-1,4))
plt.xticks(np.arange(0,6))
plt.xlim([0,t_end])
plt.ylim([-1,3])
plt.show()
```

## Plotting the Fourier decomposition

```python
# solving the three harmonic frequencies
# X = np.concatenate([np.cos(2*pi*T*np.arange(1,4)*f_best), np.sin(2*pi*T*np.arange(1,4)*f_best)],axis=1)
s1 = X_best[:,[0,3]]@c_best[[0,3]]
s2 = X_best[:,[1,4]]@c_best[[1,4]]
s3 = X_best[:,[2,5]]@c_best[[2,5]]
```

```python
fig, ax = plt.subplots(nrows=3,figsize=(8,8))
fig.tight_layout()
ax[0].plot(T,s1)
ax[1].plot(T,s2)
ax[2].plot(T,s3)
for i in range(3):
    ax[i].set_ylim([-1,1])
    ax[i].set_xlim([0,t_end])
plt.show()
```

```python
# solving the amplitudes of the three harmonics
p1 = c_best[0]**2 + c_best[3]**2
p2 = c_best[1]**2 + c_best[4]**2
p3 = c_best[2]**2 + c_best[5]**2

print(f"Amplitudes: p1={p1:.2f}, p2={p2:.2f}, p3={p3:.2f}.")
```

```python
f_vec = np.array([f_best,2*f_best,3*f_best])
p_vec = np.array([p1,p2,p3])

plt.figure(figsize=(6,6))
plt.stem(f_vec,np.sqrt(p_vec),basefmt=" ")
plt.ylim([0,1.1])
plt.ylabel("Amplitude")
plt.xlim([0,10])
plt.xlabel("Frequency [Hz]")
plt.show()
```

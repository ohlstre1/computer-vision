# Respiration belt signal example: LPF filter
We illustrate the usage of butterworth LPF on respiration data 

*ELEC-C5212* - Johdatus signaalien tilastolliseen mallintamiseen / Esa Ollila

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns
from scipy.signal import butter, freqz, lfilter, filtfilt

sns.set_context("poster")
sns.set_style("whitegrid") 
# rc('font',**{'family':'serif','serif':['Computer Modern Roman']})
rc("text",usetex=True)
rc('lines', linewidth=2.5)
```

## Loading the noisy respiration belt signal and plotting it
Sampling frequency is 
$
f_s = 1/T_s, 
$
where $T_s$ is the sampling interval (in seconds).

```python
respiration_n = np.loadtxt("resp_noisy.txt")
Ts = 0.001
fs = 1/Ts
ind = np.arange(0,len(respiration_n),100)
t = np.arange(len(respiration_n))/fs
t_end = 20
plt.figure(figsize=(10,3))
plt.plot(t[ind],respiration_n[ind],'b')
plt.xlim([0,t_end]),plt.ylim([-3,3])
plt.xlabel("Time [s]")
plt.xticks(np.arange(0,t_end+2,2))
plt.show()
```

## Defining a butterworth low-pass filter

```python
fc = 2 # cutoff frequency at 2 Hz
fc_n = fc/(fs/2) # normalized in the [0,1] range where 1 corresponds to the nyqist frequency
b,a = butter(4,fc_n,"low")
f, H = freqz(b,a,10000,fs=fs)
# plotting the gain of the filter
plt.figure(figsize=(6,6))
plt.plot(f[f<10], np.abs(H[f<10]))
plt.xlim([0,10])
plt.xlabel("Frequency [Hz]")
plt.ylim([0,1.01])
plt.ylabel("$|H(\exp(j2\pi f/f_s))|$")
plt.show()
```

## Plotting the filtered signals

Then filter the  signals. Compare `lfilter` to `filtfilt`

```python
respiration_lfilt = lfilter(b,a,respiration_n) # corresponds to matlabs filter function
respiration_filtfilt = filtfilt(b,a,respiration_n) # corresponds to matlabs filtfilt function
```

Then plot:

```python
ind = np.arange(0,8000,100) # plot 8 seconds 
fig, ax = plt.subplots(nrows=2, figsize=(10,6))
fig.tight_layout()
ax[0].plot(t[ind],respiration_n[ind],'b')
ax[1].plot(t[ind],respiration_lfilt[ind])
ax[1].plot(t[ind],respiration_filtfilt[ind])
ax[1].legend(["\\texttt{lfilter}","\\texttt{filtfilt}"])

# setting plot limits to 8 seconds of the whole signal
for i in range(2):
    ax[i].set_xlim([0,8])
    ax[i].set_ylim([-3,3])
    ax[i].set_xlabel("Time [s]")
plt.show()
```

* Stocks are securities representing equity ownership in a corporation. 
* The price or cost $C_t$ of a stock at some time $t$ is always non-negative, $0\leq C_t \leq \infty$.    
* The **net return**  of a stock over one interval (or time period) from $t-1$ to $t$ is
$$
r_t = \frac{C_t-C_{t-1}}{C_{t-1}}=\frac{C_t}{C_{t-1}}-1 \in [-1,\infty)
$$
One often refers to net return simply as the return (or profit rate) and it is normally quoted as a percentage. 

* Net returns are numbers like 0.02 (when the stock rose 2\%) or $-0.02$ (when the stock fell 2\%). Value $-1$ for a net return means bankruptcy. 

* The fraction $1+r_t=C_t/C_{t-1}\in[0,\infty)$ is called the **gross return** or total return.

## Are daily stock (net) returns  normally distributed? 
* Let us example this question empirically. 

The  Gaussianity of data can be investigated via:
* [quantile-quantile (QQ-)plot](https://en.wikipedia.org/wiki/Q%E2%80%93Q_plot)
* comparing the plots of  empirical pdfs (histrograms) against the theoretical ones. 
* computing skewness/kurtosis statistics: do they compare well with theoretical values for the Gaussian distribution?


We make an analysis of NASDAQ-100 (^NDX) stock market index returns. 

```python
# import libraries and functions 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import seaborn as sns
import pandas as pd
from scipy.stats import skew, kurtosis, probplot, norm

np.random.seed(0)

sns.set_context("poster")
sns.set_style("whitegrid") 
rc('font',**{'family':'serif','serif':['Computer Modern Roman']})
rc("text",usetex=True)
rc('lines', linewidth=2.5)
rc('lines', markersize=3.5)
```

## Loading and plotting the NDX100 data

```python
NDX100 = pd.read_csv("NDX100.txt",sep="\t")
dates = pd.to_datetime(NDX100["Date"],infer_datetime_format=True).dt.to_pydatetime()
prices = NDX100["Close"].to_numpy()
plt.figure(figsize=(10,6))
plt.plot(dates, NDX100["Close"])
plt.xlim([dates.min(), dates.max()])
plt.xticks(dates[np.arange(0,len(NDX100),80)])
plt.xlabel("Date")
plt.ylabel("Close [USD]")
plt.show()
```

```python
NDX100.head()
#NDX100.tail()
```

Then compute the gross returns and net returns

```python
gross_return = prices[1:-1]/prices[0:-2]
net_return = gross_return-1
```

## Computing sample statistics and comparing them with Gaussian data

```python
mu = np.mean(net_return)
sd = np.std(net_return)
sk = skew(net_return)
ku = kurtosis(net_return, fisher=False)
print(f"NDX100: mu {mu:.4f}, sd {sd:.4f}, sk {sk:.4f}, ku {ku:.4f}")

gaussian_data = np.random.randn(len(net_return))
mu = np.mean(gaussian_data)
sd = np.std(gaussian_data)
sk = skew(gaussian_data)
ku = kurtosis(gaussian_data, fisher=False)
print(f"gaussian: mu {mu:.4f}, sd {sd:.4f}, sk {sk:.4f}, ku {ku:.4f}")
```

**Q:** What can you infer from skewness kurtosis estimates. Are they close to Gaussian values?

## Plots of prices and returns

```python
fig, ax = plt.subplots(nrows=2, figsize=(10,8))
fig.tight_layout()
ax[0].plot(dates[0:-2],NDX100["Close"][0:-2])
ax[1].plot(dates[0:-2],np.zeros_like(NDX100["Close"])[0:-2],'--r')
ax[1].plot(dates[0:-2],net_return)
ax[0].set_ylabel("Close [USD]")
ax[1].set_ylabel("Net Return [USD]")
ax[1].set_ylim([-0.03,0.03])
for i in range(2):
    ax[i].set_xlabel("Date") 
    ax[i].set_xlim([dates[0], dates[-2]])
    ax[i].set_xticks(dates[np.arange(0,len(NDX100),80)])
plt.show()
```

## QQ-plots
Let's see if QQ-plots confirm Gaussianity of stock returns

```python
net_return_std = (net_return-net_return.mean())/net_return.std()
fig, ax = plt.subplots(ncols=2,figsize=(10,6))
fig.tight_layout()
probplot(net_return_std,plot=ax[0])
probplot(gaussian_data,plot=ax[1])
ax[1].set_ylim([-4.5,4.5])
plt.show()
```

**Q:** Compare left and right plots. Does the net returns look like being a random sample from a Gaussian distribution? 

## Histograms
We can inspect the empirical pdf of Nasdaq-100 returns to get a better picture we compare with empirical pdf of the simulated Gaussian sample from $\mathcal N(0,1)$.

```python
nbins=20
x_axis = np.arange(-5, 5, 0.001)
fig,ax = plt.subplots(ncols=2,figsize=(10,6))
ax[0].hist(net_return_std, nbins, range=[-5,5], density=True)
ax[0].plot(x_axis, norm.pdf(x_axis,0,1))
ax[0].set_title("NASDAQ-100")

ax[1].hist(gaussian_data, nbins, range=[-5,5], density=True)
ax[1].plot(x_axis, norm.pdf(x_axis,0,1))
ax[1].set_title("$\mathcal{N}(0,1)$ Sample")

for i in range(2):
    ax[i].set_xlim([-5,5])
    ax[i].set_ylim([0,0.6])
plt.show()
```

*System identification using Wiener filters*

In **system identification**, the filter is placed in parallel with the unknown system: 

A sequence of white noise $x(n)$  is given as input  to the Wiener filter and to the unknown system and  the Wiener filter coefficients are adjusted to minimize the difference between the output of the filter and of the unknown system.  The parameter to play with is the filter order. In the first part, the unknown system is as an all-zero (i.e., FIR) linear system. 

```python
N = 1024 # number of samples 
b = [0.264, 1.405, 3.331, 3.331, 1.405, 0.264] # The unknown system
np.random.seed(0)     # set seed for reproducibility 
x = np.random.randn(N) # known input to the systems 
d = lfilter(b,1.,x)   # measured output from the unknown system
```

Above `x` and `d`  are the input and desired signal, respectively,  given as input to the Wiener filter. 

### Question 2a (2P.)

* Plot the gain (magnitude of the frequency response) for the true unknown system and for the  filters.

Comment on which filter order seems to provide the best match to the unknown system? 

_solution_: P = 5 has the most lookalike curve of the wiener filters

```python
# your code 
p = [1, 5, 9]

winer_coef_1 = wiener(x, d, p[0])
winer_coef_5 = wiener(x, d, p[1])
winer_coef_9 = wiener(x, d, p[2])
w1, h1 = freqz(winer_coef_1)
w5, h5 = freqz(winer_coef_5)
w9, h9 = freqz(winer_coef_9)
w_unkown, h_unkown = freqz(b)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,10))

# your code 
ax[0][0].plot(w1/pi ,np.abs(h1))
ax[0][0].set_title("Wiener Filter order p = 1")
ax[0][0].grid(), 
ax[0][0].set_ylim([ -1,11])
ax[0][0].set_ylabel('$|W(e^{\jmath \omega})|$')

# your code 
ax[0][1].plot(w5/pi ,np.abs(h5))
ax[0][1].set_title("Wiener Filter order p = 5")
ax[0][1].grid()
ax[0][1].set_ylim([ -1,11])

# your code 
ax[1][0].plot(w9/pi ,np.abs(h9))
ax[1][0].set_title("Wiener Filter order p = 9")
ax[1][0].grid()
ax[1][0].set_ylim([ -1,11])
ax[1][0].set_ylabel("$|W(e^{\jmath \omega})|$")
ax[1][0].set_xlabel("Frequency $\omega$ (in units of $\pi$)")

# your code
ax[1][1].plot(w_unkown/pi ,np.abs(h_unkown))
ax[1][1].set_title("True unknown system")
ax[1][1].grid()
ax[1][1].set_ylim([ -1,11])
ax[1][1].set_xlabel("Frequency $\omega$ (in units of $\pi$)")

plt.show()
```

A: The required length is order 5 as the unknown system is also a  FIR system of order 5.

*Adaptive noise cancellation using the NLMS*

In Section 7.4 of lecture notes we considered noise cancellation application of the Wiener filter.  In this question we perform **adaptive noise cancellation (ANC)** using the NLMS.  The configuration of ANC is as follows: 

Compared to configuration illustrated in Figure 7.3 of lectures notes is that in ANC filter weights are adaptively estimated at each time instant, that is, an adaptive filter is used instead of the stationary Wiener filter (see also Example 7.3 in lecture notes). 

**data generation**

$x(n)=d(n)+v_1(n)$ are the obtained measurements at the primary sensor, where 
* desired signal $d(n)$ is a sine wave  with 2 Hz frequency for the first 3 seconds and 4 Hz frequency for the next 3 seconds. 
* noise $v_1(n)$ at primary sensor is a 5.0 Hz sawtooth waveform. 

The secondary sensor measures reference signal $v_2(n)$, which (in this example) is generated as a low pass filtered and phase shifted version of  $v_1(n)$.  

```python
fs = 500         # Sampling frequency 
N  = 3000        # # of samples (6 seconds)
t = np.arange(N)/fs # time instants for plotting

# Generate the signal from primary sensor 
d = np.zeros(N)
t = np.arange(N)/fs # time instants 
d[:int(N/2)] = np.cos(t[:int(N/2)] * 2 * pi * 2) # 2Hz  signal
d[int(N/2):] = np.cos(t[int(N/2):] * 2 * pi * 4) # % 4Hz signal
```

```python
from scipy.signal import sawtooth
v1 = sawtooth(2 * pi * 5 *t,0.5)
x = d + v1      # measured signal at primary sensor 

# Generate the reference signal from secondary sensor
wn = 8/(fs/2) # Cutoff frequency at 8 Hz
b,a = butter(4,wn,'low') # lowpass filtered signal
v2 = lfilter(b,a,v1)
a
```

Here are the plots of the signal $d(n)$, measured signal $x(n)$  and the reference signal $v_2(n)$. 

```python
fig, ax = plt.subplots(nrows=3,figsize=(14,9))
ax[0].plot(t, d)
ax[0].set_title("signal",fontsize=14)
ax[0].grid()
ax[0].set_ylabel("$d(n)$")

ax[1].plot(t,x)
ax[1].set_title("Measured (noisy) signal",fontsize=14)
ax[1].grid()
ax[1].set_ylabel("$x(n)$")

ax[2].plot(t,v2)
ax[2].set_title("Reference signal",fontsize=14)
ax[2].grid()
ax[2].set_ylabel("$v_2(n)$")
plt.plot()
```

We consider the spatial filter $ y(n)= \mathbf{w}^\top \mathbf{x}(n)$, where $\mathbf{x}(n)$  is an output of an array model 
$$ 
\mathbf{x}(n)= \mathbf{a} s(n) +  \mathbf{e}(n)
$$
where $\mathbf{a} \in  \mathbb{C}^p$ is the array response and  $\mathbf{e}(n) \in \mathbb{C}^p$ is a zero-mean spatially and temporally white Gaussian noise process, uncorrelated with the signal $s(n)$. Thus $\mathrm{E}[\|\mathbf{e}(n)\|^2] = p \sigma^2$, where $\mathrm{var}(e_i(n))=\sigma^2$.  The SNR is defined by 
$$
\mathrm{SNR} = 10 \log_{10} \frac{|s(n)|^2\| \mathbf{a} \|^2}{\sigma^2 p} 
$$
and we assume that SNR = 15 dB. 

The signal is 8-PSK (Phase-shift keying signal with modulation order $M=8$) and hence the signal has constant modulus $|s(n) | = 1$.  We generate $N=1000$ samples from such an array model in the case that the number of sensors in the array is $p=10$.

**Generate the data**

### Question 5b (2P.)
Recall from lecture notes that CMA algorithm is 
$$
\mathbf{w}_{n+1} =  \mathbf{w}_n   +  \mu    ( 1- | y(n) |^2 \big)   y(n)  \mathbf{x}(n)^*     ,
$$
where $y(n) = \mathbf{w}^\top_n \mathbf{x}(n)$  is the spatial filter output at time $n$. 

Implement the CMA algorithm in the function called `cma_update(x,mu,w_n)` below which takes as its arguments $\mathbf{x}(n), \mu $ and $\mathbf{w}_n$ and computes the update  using the formula above. 

```python
def cma_update(x,mu,w_old):
    """ Updates the CMA weight wector given the array measurement p x 1 vector x(n).
    The step size is 'mu' and the previous p x 1 weight vector is given by w_old """
    # some code 
    y = np.transpose(w_old)@x
    w_plus_one = w_old + mu*(1-np.abs(y)**2)*y*np.conj(x)
    return w_plus_one
```

## Question 6 (5P)<a id='q6'></a>
We generate a $N=2^7=128$ samples from the model:
$$
x(n) = A \cos( \omega_0 n) + v(n),  n = 0,1,\ldots, N
$$
with $\omega_0 =0.1068$, $A = 0.67$,  and $v(n) \sim {\cal N}(0,0.25)$.

```python
N = 2**7
A = 0.67
om0 = 0.1068
np.random.seed(1234)
v = np.sqrt(0.25) * np.random.randn(N) # this is measurement noise 
n = np.arange(N)
sinusoid = A*np.cos(om0*n) # transmitted signal 
x = sinusoid + v # our measured signal 
```

### Question 6a (3P.)
Find the angular frequency $\omega_0$ by 
$$
\hat{\omega}_0 = \text{value of } \omega \in [0,\pi) \text{ that maximizes } \left|\sum_{n=0}^{N-1} x(n) e^{- \jmath \omega n}\right|.
$$
that is, your finding the **frequency that maximizes the (1-sided) periodogram** over $\omega$. 

Your task is to implement a brute force grid search method for finding the maximum $\hat \omega_0$. Note: this is straightforward by noting that the sum expression is in fact a dot product and we may the problem as:

$$
\hat{\omega}_0 = \text{value of } \omega \in [0, \pi) \text{ that maximizes } \left|\mathbf{x} \cdot \mathbf{a}(\omega) \right|,
$$

where $\mathbf{x}=(x(0),x(1),\ldots, x(N-1))^\top$ and $\mathbf{a}(\omega)=(e^{- \jmath \omega \cdot 0}, e^{- \jmath \omega \cdot 1},\ldots , e^{-\jmath \omega \cdot (N-1)})^\top$. Use a  dense grid of $\omega$ values `omegas = np.linspace(0, pi, 1000)`.  Note: Imaginary unit in python is `1j`

```python
omegas = np.linspace(0, pi, 1000)  #  use this dense grid of frequencies
#omegas = np.linspace(0, pi, 2**6+1) # this is the grid in periodogram

# your code
max_om = 0
max_variable = 0
for om in omegas:
    # Create vector a. Assume data is in x.
    # compute | <x,a(om)> | 
    # Imaginary unit is 1j
    a = np.exp(-1j*om*n)
    variable = abs(np.dot(x,a))
    if  max_variable < variable:
        max_variable =  variable
        max_om = om
om_diff = abs(max_om-om0)
print("Omega that maximizes the (1-sided) periodogram is {:.10f}".format(max_om))
print("Difference between true value and max: {:.10f}".format(om_diff))
# your code 
```

### Question 6b (1P.)
Compute an estimate $\hat \omega_0$ as the maximum angular frequency obtained in the periodogram (use function `periodogram`). Which give more accurate answers? 

**Note:**  The (1-sided) periodogram is computed via Fast Fourier Transform (FFT) at discretized frequencies:
$$
\omega_k =  2 \pi \frac{k}{N_{\mathrm{FFT}}} ,  \ k=0,1, \ldots, N_{\mathrm{FFT}}/2
$$
where $N_{\text{FFT}} = 2^{\lceil\log_2(N)\rceil}$ is the FFT window length.  In our case $N_{FFT}=N$ since $N$ is power of $2$. Thus the grid used by the `periodogram` is not as dense as you used in a) part. 

**Note**: Also, if you choose the grid as  `omegas = np.linspace(0, pi, 2**6+1)` in a)-part, then it should match with your answer in b)-part. 

```python
f, Pxx = periodogram(x, fs = 2*pi, nfft = N)
index = np.argmax(Pxx)
estimated_om = f[index]
print("Estimated omega with periodogram: {:.10f}".format(estimated_om ))
print("Difference between true value and estimation: {:.10f}".format(abs(om0-estimated_om )))
print("Previous method more accurate")
```
